```python
def opt_func(trc_func,t_series,p_series,mu_sigma,num_bins,*faux_args):
        mu,sigma = mu_sigma
        return brwnnexp.kscomp(trc_func,t_series,p_series,0,mu,sigma,num_bins,*faux_args)[0]
```

```python
knwn_mu,knwn_sigma = (0.2,0.3)
knwn_t,knwn_w,knwn_p = brwnnexp.astep(1000,0.01,knwn_mu,knwn_sigma)

mu_vals = np.linspace(-10,10,30)
sigma_vals = np.linspace(0.2,0.4,30)
ks_space = np.array([opt_func(brwnnexp.atrace,knwn_t,knwn_p,(m,s),100) for m in mu_vals for s in sigma_vals])

space = np.array([(m,s) for m in mu_vals for s in sigma_vals])
mu_space,sigma_space = zip(*space)
```

```python
import scipy.optimize as optmz
```

```python
def optmz_sigma(trc_func,t_series,p_series,prms_init,num_bins,*faux_args):
    '''Optimize the parameter sigma of the trace function
    
    trc_func = Use etrace or atrace to act on dat_series and t_series
    t_series = The time series corrisponding to dat_series
    p_series = The data set which will be traced
    prms_init = Initial guess for the parameters mu and sigma
    num_bins = The number bins used to histogram p_series
    faux_args = Auxilary arguments for trc_func
    '''
    def opt_func(mu_sigma):
        mu,sigma = mu_sigma
        return brwnnexp.kscomp(trc_func,t_series,p_series,0,mu,sigma,num_bins,*faux_args)[0]
    
    return optmz.fmin_powell(opt_func,prms_init)
```

```python
knwn_mu,knwn_sigma = (0.2,0.3)
knwn_t,knwn_w,knwn_p = brwnnexp.astep(1000,0.01,knwn_mu,knwn_sigma)
```
