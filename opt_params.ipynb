```python
import numpy as np
import pandas as pd
from scipy.stats import ks_2samp
from datetime import datetime, timedelta
```

```python
import brownian.wiener
import quantquoteminutely as qqm
import brownian.exp as brwnnexp
import utility_funcs as ufuncs
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
aapl = qqm.get_minutely_data('aapl','20130102')
t_series = pd.Series(ufuncs.get_tseries(aapl.index))[:400]
p_series = aapl['close'][:400]
```

---
#Expectation Method
+ ###ecomp_params: uses the Euler DE

```python
e_mu,e_sigma = brwnnexp.ecomp_params(t_series,p_series)
e_optw = brwnnexp.etrace('price',t_series,p_series,0,e_mu,e_sigma)
```

```python
plt.figure(figsize=(10,2))
plt.title('Price Series')
plt.plot(t_series,p_series)
plt.show()
plt.figure(figsize=(10,2))
plt.title('Traced Wiener Series')
plt.plot(t_series,e_optw)
plt.show()
```

```python
print 'mu:', e_mu, '-- sig:', e_sigma
```

+ ###acomp_params: uses the analytical solution

```python
a_mu,a_sigma = brwnnexp.acomp_params(t_series,p_series)
a_optw = brwnnexp.atrace('price',t_series,p_series,0,a_mu,a_sigma)
```

```python
plt.figure(figsize=(10,2))
plt.title('Price Series')
plt.plot(t_series,p_series)
plt.show()
plt.figure(figsize=(10,2))
plt.title('Traced Wiener Series')
plt.plot(t_series,a_optw)
plt.show()
```

```python
print 'mu:', a_mu, '-- sig:', a_sigma
```

##1. real data comparison between the Wiener fds and a scaled normal dist. 

```python
ks_eval,bin_mkrs,trc_hist,norm_hist = brwnnexp.kscomp(brwnnexp.etrace,t_series,p_series,0,e_mu,e_sigma,100)
print 'euler trace'
print '-----------'
print 'ks statistic:', ks_eval[0]
print 'p-value:', ks_eval[1]
print
ks_eval,bin_mkrs,trc_hist,norm_hist = brwnnexp.kscomp(brwnnexp.atrace,t_series,p_series,0,e_mu,e_sigma,100)
print 'analytical trace'
print '----------------'
print 'ks statistic:', ks_eval[0]
print 'p-value:', ks_eval[1]
```

```python
plt.plot(bin_mkrs,norm_hist)
plt.plot(bin_mkrs,trc_hist)
```

#Optimization Method (work in progress)

```python
import scipy.optimize as optmz
```

```python
def optmz_params(trc_func,t_series,p_series,prms_init,num_bins,*faux_args):
    '''Optomize the parameters of the trace function
    
    trc_func = Use etrace or atrace to act on dat_series and t_series
    t_series = The time series corrisponding to dat_series
    p_series = The data set which will be traced
    prms_init = Initial guess for the parameters mu and sigma
    num_bins = The number bins used to histogram p_series
    faux_args = Auxilary arguments for trc_func
    '''
    def opt_func(mu_sigma):
        mu,sigma = mu_sigma
        return brwnnexp.kscomp(trc_func,t_series,p_series,0,mu,sigma,num_bins,*faux_args)[0][0] #could be [0][0] or [0][1]
    
    return optmz.fmin_powell(opt_func,prms_init)
```

```python
mu,sigma = optmz_params(brwnnexp.atrace,t_series,p_series,(a_mu,a_sigma),10)
print (mu,sigma)
optw = brwnnexp.atrace('price',t_series,p_series,0,mu,sigma)
```

```python
plt.figure(figsize=(10,2))
plt.title('Price Series')
plt.plot(t_series,p_series)
plt.show()
plt.figure(figsize=(10,2))
plt.title('Traced Wiener Series')
plt.plot(t_series,optw)
plt.show()
```

```python
ks_eval,bin_mkrs,trc_hist,norm_hist = brwnnexp.kscomp(brwnnexp.atrace,t_series,p_series,0,mu,sigma,100)
print 'ks statistic:', ks_eval[0]
print 'p-value:', ks_eval[1]
```

```python
plt.plot(bin_mkrs,norm_hist)
plt.plot(bin_mkrs,trc_hist)
```
