```python
import random
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
%matplotlib inline
import numpy as np
```

The methods in `exponential_brownian` implament equation (1) from [Didier Sornette's paper](http://arxiv.org/pdf/cond-mat/0104341v2.pdf):

>In the Stratonovich representation usually practiced by physicists, our fundamental stochastic
dynamical equation for the bubble price B(t) is of the form
$$
\frac{dB}{dt} = (a\mu_0+b\eta)B^m
$$

>where a and b are two positive constants and $\eta$ is a delta-correlated Gaussian white noise. The nonlinearity
Bm with exponent $m>1$ creates a singularity at some finite time $t^*$
and the multiplicative noise turns 
out to make t* stochastic (i.e., dependent upon the realization of the noise). As we shall show in more
details below, model (1) with $m=1$ is the standard geometrical Brownian motion used for describing
financial time series at a first-order of approximation.

It should be noted that the weiner process is synonymous with the case of m=0 and a=0 assuming $b=\sqrt{h}$
where the total time $t$ is split into $n$ small intervals of length and $h=t/n$. In the particular case implameted here
$a$ and $b$ can be taken to be approximately $0.005$ where $t=1 \text{ month}$ of $30$ days which is split into intervals of minutes.
Of course $a$ or $b$ can be altered slightly to see how each influence the results.

```python
class exponential_brownian:
    '''generates data using equation (1) from Sornette paper.
    m=1 is supposed to be equivalent to brownian motion
    '''
    def __init__(self, _m, _a, _b, _mu, _sigma):
        self.m = _m
        self.a = _a
        self.b = _b
        self.mu = _mu
        self.sigma = _sigma
        self.bval = 1.0
        self.bval_last = 1.0
        
        self.mu0_tot = 0
        self.time_steps = 0
    
    def get_value(self):
        
        nu = random.gauss(self.mu,self.sigma)
        mu0 = (self.bval-self.bval_last)/self.bval_last
        
        self.bval_last = self.bval
        alph = self.a*mu0
        beta = self.b*nu
        self.bval += (alph+beta)*self.bval**(self.m)
        
        self.mu0_tot += mu0
        self.time_steps += 1
        
        return [self.bval,alph,beta]
    
    def get_consts(self):
        return [self.m,self.a,mu0,self.b]
    
    def get_info(self):
        return [self.time_steps, self.mu0_tot]
```

`stock_history` simply iterates over a specified time period to generate minute data using the methods in `exponential_brownian`. In addition to this, a plotting function is included which makes it easy to plot selected portions of the data along with making scaling adjustments.

```python
class stock_history:
    
    def __init__(self, months, t_step, *params):
        
        self.t_step = t_step
        self.months = months
        self.B = exponential_brownian(*params)
        if t_step!='minute':
            self.times = []
            self.stock_dat = []
        try:
            alltimes = range(43200*months)
            if t_step=='minute':
                self.times = alltimes
                self.stock_dat = [self.B.get_value() for minute in self.times]
            else:
                if t_step=='hour':
                    unit=60
                if t_step=='day':
                    unit=1440
                for t in alltimes:
                    vals = self.B.get_value()
                    if t%unit==0:
                        self.times.append(t)
                        self.stock_dat.append(vals)
            self.stock_dat = [self.B.get_value() for minute in self.times]
            self.stock_dat = zip(*self.stock_dat)
            if np.isnan(max(self.stock_dat[1])):
                raise OverflowError
            print 'max/min return rates (rate/min):',round(max(self.stock_dat[1]),6),'/',round(min(self.stock_dat[1]),6)
        except OverflowError:
            raise ParamError('environment parameters may be unstable. reset them or try again.')
    
    def plot_history(self, interval=None, xlims=None, converge=0, contrast=1, legend_loc = 'best', savefile = None):
        '''interval = the time between the plotted data points
        xlims = data within the specified time period will be plotted (float iterable, should be given in months)
        converge = moves the dB/dt and the gaussian white noise up or down by the specified amount
        contrast = larger/smaller numbers increase/decrease the relative size of the data compared to dB/dt and the white noise
        legend_loc = provide a keyword for loc arg in pyplot.legend()
        savefile = if provided will save the figure under the specified file name
        '''
        
        def conv_unit(inter):
            if inter == 'minute':
                c = 43200
                u=1
            if inter == 'hour':
                c = 720
                u=60
            if inter == 'day':
                c = 30
                u=1440
            return c,u
        
        x= self.times
        if interval==None:
            interval = self.t_step
            conv,unit = conv_unit(interval)
            y0 = np.array(self.stock_dat[0])
            y1 = np.array(self.stock_dat[1])
            y2 = np.array(self.stock_dat[2])
        
        else:
            xn = []
            y0 = []
            y1 = []
            y2 = []
            conv,unit = conv_unit(interval)
            for t in x:
                if t%unit==0:
                    xn.append(t)
                    y0.append(self.stock_dat[0][t])
                    y1.append(self.stock_dat[1][t])
                    y2.append(self.stock_dat[2][t])
            x = np.array(xn)
            y0 = np.array(y0)
            y1 = np.array(y1)
            y2 = np.array(y2)
        
        if xlims==None:
            widen = 1
        else:
            tmin = xlims[0]
            tmax = xlims[1]
            widen = (float(self.months)/(tmax-tmin))**0.3
            x= self.times[int(tmin*conv):int(tmax*conv)]
            y0 = np.array(y0[int(tmin*conv):int(tmax*conv)])
            y1 = np.array(y1[int(tmin*conv):int(tmax*conv)])
            y2 = np.array(y2[int(tmin*conv):int(tmax*conv)])
        
        my0 = max(abs(y0))
        my1 = max(abs(y1))
        my2 = max(abs(y2))
        
        contrast*=2
        y0 = y0*contrast+min(y0)*(1-contrast)
        y1 = y1*my0/my1/2/contrast+converge
        y2 = y2*my0/my2/2/contrast+converge
        
        my1 = max(abs(y1))
        my2 = max(abs(y2))
        
        fig,ax = plt.subplots(figsize = (20,10))
        params = {'legend.fontsize': 14,
                  'legend.linewidth': 1.5,
                  'legend.markerscale': 1}
        plt.rcParams.update(params)
        
        plt.plot(x,y0, 'k-', linewidth = 3.5/conv**0.15)
        plt.plot(x,y2, 'r-', linewidth = 1.5/conv**0.4*widen)
        plt.plot(x,y1, 'b-', linewidth = 1.0/conv**0.4*widen)
        
        all_dat = np.concatenate([y0,y1,y2])
        if contrast==2:
            is_scaled = ''
        else:
            is_scaled = ' scaled'
        plt.ylim(0.95*min(all_dat), 1.05*max(all_dat))
        plt.plot(0,0,'k-', label = 'Stock Price (B{0})'.format(is_scaled))
        plt.plot(0,0,'b-', label = 'dB/dt (scaled)')
        plt.plot(0,0,'r-', label = 'Gaussian White Noise (scaled)')
        plt.legend(loc = legend_loc)
        
        
        
        plt.yticks(fontsize = 17)
        plt.xticks(fontsize = 17)
        
        if xlims==None:
            majorLocator = MultipleLocator(43200)
            minorLocator = MultipleLocator(10800)
            ax.xaxis.set_major_locator(majorLocator)
            ax.xaxis.set_minor_locator(minorLocator)
            ax.set_xticklabels(range(-1,self.months+1))
            plt.xlim(xmax = x[-1])
        else:
            tdif = tmax-tmin
            if tdif<1:
                stp = float(tdif)/4
                ax.set_xticks(np.arange(int(tmin*conv),int(tmax*conv)+1,stp*conv))
                ax.set_xticks(np.arange(int(tmin*conv),int(tmax*conv)+1,stp*conv/4),minor=True)
                ax.set_xticklabels(np.arange(tmin,tmax+stp,stp))
                
            
            else:
                ax.set_xticklabels(np.arange(tmin,tmax+1,1))
                ax.set_xticks(np.arange(int(tmin*conv),int(tmax*conv)+1,conv))
                ax.set_xticks(np.arange(int(tmin*conv),int(tmax*conv)+1,float(conv)/4),minor=True)
                
            plt.xlim(int(tmin*conv),int(tmax*conv))
        
        ax.xaxis.set_tick_params(which='major', size=15, width=1.5)
        ax.xaxis.set_tick_params(which='minor', size=5, width=1.5)
        plt.xlabel('Months', fontsize = 25)
        
        plt.grid(which='major',linewidth=1.0)
        plt.grid(which='minor',linewidth=0.35)
        
        if savefile!=None:
            plt.savefig(savefile, bbox_inches = 'tight')
        
        plt.show()
```

```python
#Suggested parameters: [2,0.005,0.005,0,1]
#The above parameters may not remain stable for extended periods of time
#See the saved_output directory for example output
def gen_bs_data(it_max,t_period,*params):
    
    class ParamError(OverflowError):
        pass
    
    count = 0
    while True:
        try:
            dat = stock_history(t_period,'minute',*params)
            break
        except ParamError:
            print 'Error count:', count
            count+=1
            if count>it_max:
                print 'exceeded the maximum number of iterations: returned no data.'
                return None
            pass
    return dat
```

```python
class exponential_brownian:
    '''Generates data using equation (1) from Sornette paper.
    m=1 is supposed to be equivalent to brownian motion
    '''
    
    class FillError(Exception):
        def __init__(self, kind):
            self.string = ('The {0} series has not been defined. '
                           "Pass it one in the class constructor through **kwargs or arg 'x' in retrace()").format(kind)
        def __str__(self):
            return str(self.string)
    
    def __init__(self, m, a, b, mu, sigma, **kwargs):
        '''Provide the parameters included in the differential equation above.
        **kwargs makes it possible to initialize with a preexisting price or weiner series.'''
        
        self.m = m
        self.a = a
        self.b = b
        self.mu = mu
        self.sigma = sigma
        
        self.init_pval_last = kwargs.get('pval_last',1.0)
        self.init_pval_crnt = kwargs.get('pval_crnt',1.0)
        
        self.pval_last = self.init_pval_last
        self.pval_crnt = self.init_pval_crnt
        self.p_series = kwargs.get('p_series',[self.pval_crnt])
        self.w_series = kwargs.get('w_series',[0])
    
    def step(self, number):
        '''Moves the price series forward through the specified number of steps.'''
        
        for i in range(number):
            nu = random.gauss(self.mu,self.sigma)
            self.w_series.append(nu)
            mu0 = (self.pval_crnt-self.pval_last)/self.pval_last

            self.pval_last = self.pval_crnt
            alph = self.a*mu0
            beta = self.b*nu
            
            self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
            self.p_series.append(self.pval_crnt)
    
    def retrace(self, series, x=None, **kwargs):
        '''Overrides the current price/weiner series with a new one based on a retrace of a weiner/price series.
        
        series = 'weiner': 
        Retraces the values of the weiner series to create a price series.
        series = 'price':
        Retraces the values of the price series to create a weiner series.
        
        If x != None then x is overrides preexisting series data and is
        considered to be the series which is being retraced.
        NOTE: the provided series must have more than one value.'''
        
        if series == 'weiner':
            self.pval_last = self.init_pval_last
            self.pval_crnt = self.init_pval_crnt
            self.p_series = [self.pval_last]
            if x != None:
                self.w_series = x
            elif self.w_series == []:
                raise exponential_brownian.FillError(series)
            for nu in self.w_series[1:]:
                mu0 = (self.pval_crnt-self.pval_last)/self.pval_last
                self.pval_last = self.pval_crnt
                alph = self.a*mu0
                beta = self.b*nu

                self.pval_crnt += (alph+beta)*self.pval_crnt**(self.m)
                self.p_series.append(self.pval_crnt)
        
        elif series == 'price':
            self.w_series = [0]
            
            if x != None:
                self.p_series = x
            elif self.p_series == []:
                raise exponential_brownian.FillError(series)
            
            pval_last = self.init_pval_last
            pval_crnt = self.init_pval_crnt
            
            for i in range(1,len(self.p_series)):

                pval_next = self.p_series[i]
                
                dP = (pval_next - pval_crnt)/(pval_crnt**(self.m))
                mu0 = (pval_crnt - pval_last)/pval_last
                self.w_series.append((dP-self.a*mu0)/self.b)
                
                pval_last = pval_crnt
                pval_crnt = pval_next
```
